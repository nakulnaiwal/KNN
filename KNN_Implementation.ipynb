```python
import numpy as np
import pandas as pd
import math
from collections import Counter
```

```python
class KNN:
  def __init__ (self, k=3):
    self.K = k
    self.X_train = None
    self.Y_train = None

  def fit(self, X,Y):
    self.X_train = X
    self.Y_train = Y

  def euclidean(self,a, b):
    assert len(a) == len(b)

    dist = 0
    for i in range((len(a))):
      diff = (a[i] - b[i])
      dist += diff**2

    return math.sqrt(dist)

  def predict_one(self, x):
    distances = []

    for i in range(len(self.X_train)):
      dist = self.euclidean(x,self.X_train[i])
      distances.append((dist,self.Y_train[i])) # Fixed: append a tuple


    distances.sort(key=lambda pair:pair[0])

    k_nearest = distances[:self.K]

    labels = [label for (_, label) in k_nearest]
    most_common = Counter(labels).most_common(1)
    return most_common[0][0]

  def predict(self, X_test):
    predictions = []
    for x in X_test:
      predictions.append(self.predict_one(x))
    return predictions
```

```python
X_train = [
    [1, 2],
    [2, 3],
    [3, 3],
    [6, 7],
    [7, 8],
    [8, 8]
]

y_train = [0, 0, 0, 1, 1, 1]

knn_model = KNN()
knn_model.fit(X_train, y_train)
ans = knn_model.predict([[5,5]])
print(ans)
```
